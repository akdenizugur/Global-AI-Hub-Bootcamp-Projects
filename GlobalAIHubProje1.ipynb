```python
import csv
import pandas as pd
```

```python
Ogrenciler = []
OgrenciSayisi = 0
```

```python
class Sinif:
    def KayitEkle():
        Vize = 0
        Final = 0
        global Ogrenciler,OgrenciSayisi
        durum = 1
        OgrenciNumarası = int(input("Öğrenci No : "))
        for i in Ogrenciler:
            if(OgrenciNumarası == i[0]):
                print("Numara Sisteme Kayıtlı")
                durum = 0
                break
            else:
                durum = 1
        if durum:
            while(True):
                OgrenciAdi = input("Ad : ")
                OgrenciSoyadi = input("Soyad : ")
                DersIsmi = input("Ders İsmi : ")
                Vize = int(input("Vize : "))
                if Vize > 100 or Vize < 0:
                    print("Lütfen Geçerli Bir Not Giriniz.")
                    break
                Final = int(input("Final : "))
                if Final > 100 or Final < 0:
                    print("Lütfen Geçerli Bir Not Giriniz.")
                    break
                liste = [int(OgrenciNumarası), OgrenciAdi, OgrenciSoyadi, DersIsmi, Vize, Final]
                Ogrenciler.append(liste)
                OgrenciSayisi += 1
                print(OgrenciNumarası," Numaralı Öğrenci Kayıt Edildi.")
                

                VizePuani = float((Vize * 40.0) / 100.0)
                FinalPuani = float((Final * 60.0) / 100.0)
                GectiKaldi = ""
                Durum = ""
                Ortalama = float(VizePuani + FinalPuani)
                liste.append(round(Ortalama,2))

                if Ortalama <= 100.0 and Ortalama >= 87.6:
                    Durum = "AA"
                    GectiKaldi = "Geçti"
                    liste.append(Durum)
                    liste.append(GectiKaldi)
                elif Ortalama <= 87.5 and Ortalama >= 81.6:
                    Durum = "BA"
                    GectiKaldi = "Geçti"
                    liste.append(Durum)
                    liste.append(GectiKaldi)
                elif Ortalama <= 81.5 and Ortalama >= 75.6:
                    Durum = "BB"
                    GectiKaldi = "Geçti"
                    liste.append(Durum)
                    liste.append(GectiKaldi)
                elif Ortalama <= 75.5 and Ortalama >= 67.6:
                    Durum = "CB"
                    GectiKaldi = "Geçti"
                    liste.append(Durum)
                    liste.append(GectiKaldi)
                elif Ortalama <= 67.5 and Ortalama >= 59.6:
                    Durum = "CC"
                    GectiKaldi = "Geçti"
                    liste.append(Durum)
                    liste.append(GectiKaldi)
                elif Ortalama <= 59.5 and Ortalama >= 49.6:
                    Durum = "DC"
                    GectiKaldi = "Kaldı"
                    liste.append(Durum)
                    liste.append(GectiKaldi)
                elif Ortalama <= 49.5 and Ortalama >= 39.6:
                    Durum = "DD"
                    GectiKaldi = "Kaldı"
                    liste.append(Durum)
                    liste.append(GectiKaldi)
                elif Ortalama <= 39.5 and Ortalama >= 29.5:
                    Durum = "FD"
                    GectiKaldi = "Kaldı"
                    liste.append(Durum)
                    liste.append(GectiKaldi)
                else:
                    Durum = "FF"
                    GectiKaldi = "Kaldı"
                    liste.append(Durum)
                    liste.append(GectiKaldi)
                
                break
    
        
    def KayitSil():
        global Ogrenciler, OgrenciSayisi
        sil = int(input("Silmek İstediğiniz Kayıtın Numarasını Girin : "))
        for i, eleman in enumerate(Ogrenciler):
            if(eleman[0] == sil):
                Ogrenciler.pop(i)
                OgrenciSayisi -= 1
                print("Kayıt Silindi : ", sil)
                break
            else:
                print("Kayıt Yok : ", sil)
         
    def KayitGuncelle():
        global Ogrenciler, OgrenciSayisi
        guncelle = int(input("Güncellemek İstediğiniz Kaydın Numarasını Girin : "))
        for i, eleman in enumerate(Ogrenciler):
            if(eleman[0] == guncelle):
                while(True):
                    print("Bilgileri Tekrar Girin : ", guncelle)
                    OgrenciNumarası = int(input("Öğrenci No : "))
                    OgrenciAdi = input("Ad : ")
                    OgrenciSoyadi = input("Soyad : ")
                    DersIsmi = input("Ders İsmi : ")
                    Vize = int(input("Vize : "))
                    if Vize > 100 or Vize < 0:
                        print("Lütfen Geçerli Bir Not Giriniz.")
                        break
                    Final = int(input("Final : "))
                    if Final > 100 or Final < 0:
                        print("Lütfen Geçerli Bir Not Giriniz.")
                        break
                    Ogrenciler[i] = [int(OgrenciNumarası), OgrenciAdi, OgrenciSoyadi, DersIsmi, Vize, Final]

                    VizePuani = float((Vize * 40.0) / 100.0)
                    FinalPuani = float((Final * 60.0) / 100.0)
                    GectiKaldi = ""
                    Durum = ""
                    Ortalama = float(VizePuani + FinalPuani)
                    Ogrenciler[i].append(round(Ortalama,2))

                    if Ortalama <= 100.0 and Ortalama >= 87.6:
                        Durum = "AA"
                        GectiKaldi = "Geçti"
                        Ogrenciler[i].append(Durum)
                        Ogrenciler[i].append(GectiKaldi)
                    elif Ortalama <= 87.5 and Ortalama >= 81.6:
                        Durum = "BA"
                        GectiKaldi = "Geçti"
                        Ogrenciler[i].append(Durum)
                        Ogrenciler[i].append(GectiKaldi)
                    elif Ortalama <= 81.5 and Ortalama >= 75.6:
                        Durum = "BB"
                        GectiKaldi = "Geçti"
                        Ogrenciler[i].append(Durum)
                        Ogrenciler[i].append(GectiKaldi)
                    elif Ortalama <= 75.5 and Ortalama >= 67.6:
                        Durum = "CB"
                        GectiKaldi = "Geçti"
                        Ogrenciler[i].append(Durum)
                        Ogrenciler[i].append(GectiKaldi)
                    elif Ortalama <= 67.5 and Ortalama >= 59.6:
                        Durum = "CC"
                        GectiKaldi = "Geçti"
                        Ogrenciler[i].append(Durum)
                        Ogrenciler[i].append(GectiKaldi)
                    elif Ortalama <= 59.5 and Ortalama >= 49.6:
                        Durum = "DC"
                        GectiKaldi = "Kaldı"
                        Ogrenciler[i].append(Durum)
                        Ogrenciler[i].append(GectiKaldi)
                    elif Ortalama <= 49.5 and Ortalama >= 39.6:
                        Durum = "DD"
                        GectiKaldi = "Kaldı"
                        Ogrenciler[i].append(Durum)
                        Ogrenciler[i].append(GectiKaldi)
                    elif Ortalama <= 39.5 and Ortalama >= 29.5:
                        Durum = "FD"
                        GectiKaldi = "Kaldı"
                        Ogrenciler[i].append(Durum)
                        Ogrenciler[i].append(GectiKaldi)
                    else:
                        Durum = "FF"
                        GectiKaldi = "Kaldı"
                        Ogrenciler[i].append(Durum)
                        Ogrenciler[i].append(GectiKaldi)
                    print("Kayıt Güncellendi : ", guncelle)
                    break
                else:
                    print("Kayıt Yok : ", guncelle)
                    break
                
    def TumKayıtlariSil():
        Ogrenciler.clear()
        dosya = open("Notlar.csv","r+")
        dosya.truncate(0)
        dosya.close()
        print("Tüm Kayılar Silindi")
        
    def Yaz():
        global Ogrenciler, OgrenciSayisi
        col = ["Öğrenci Numarası","Öğrenci Adı","Öğrenci Soyadı","Ders İsmi","Vize Notu","Final Notu","Ortalama Notu","Durum","Geçti - Kaldı"]
        rows = Ogrenciler
        
        with open('Notlar.csv', 'a', newline='', encoding='UTF-8') as f:
            write = csv.writer(f)
            #write.writerow(col)
            write.writerows(rows)
            OgrenciSayisi += 1
        f.close()
                
    def Oku():
        with open('Notlar.csv', encoding='UTF-8') as f:
            reader = csv.reader(f)
            for row in reader:
                global Ogrenciler,OgrenciSayisi
                print(row)
        f.close()
        
        
    def Csv():  
        global Ogrenciler,OgrenciSayisi
        dataread = pd.read_csv('Notlar.csv')
        Ogrenciler = dataread[['Öğrenci Numarası','Öğrenci Adı','Öğrenci Soyadı','Ders İsmi','Vize Notu','Final Notu','Ortalama Notu','Durum' , 'Geçti - Kaldı']].values.tolist()
        
    def CsvToExcel():
        excel = pd.read_csv ('Notlar.csv')
        excel.to_excel ('Notlar.xlsx', index = None, header=True)
```

```python
readKontrol = 1
def writeMenu():
    global readKontrol,OgrenciSayisi,Ogrenciler
    while(True):
        print("1 - Kayıtları Csv Dosyasına Yaz / Kayıt İşlemlerinden Sonra Bu Seçeneği Kullanın")
        print("2 - Kayıtları Csv Dosyasından Oku")
        print("3 - Yeni Kayıt Ekle")
        print("4 - Kayıt Sil")
        print("5 - Kayıt Güncelle")
        print("6 - Tüm Kayıtları Sil")
        print("7 - Csv Dosyasından Kayıtları Yükle")
        print("8 - Csv Dosyasını Excele Dönüştür ve Kaydet")
        print("9 - Çıkış")
        islem = int(input("Lütfen Bir İşlem Seçiniz : "))
        return islem
```

```python
while(True):
    islem = writeMenu()

    if islem == 1:
        Sinif.Yaz()

    elif islem == 2:
        if readKontrol == 1:
            Sinif.Oku()
            readKontrol = 0

    elif islem == 3:
        if readKontrol == 1:
            Sinif.KayitEkle()
        else:
            break
    
    elif islem == 4:
        if readKontrol == 1:
            Sinif.KayitSil()
        else:
            break
            
    elif islem == 5:
        if readKontrol == 1:
            Sinif.KayitGuncelle()
        else:
            break
    elif islem == 6:
        Sinif.TumKayıtlariSil()
    
    elif islem == 7:
        Sinif.Csv()
    
    elif islem == 8:
        Sinif.CsvToExcel()
    
    elif islem == 9:
        break
```
